```python
import random

def rock_paper_scissors():
    choices = ['rock', 'paper', 'scissor']
    while True:
        my_choice = input("Enter your choice (rock, paper, scissor) or 'quit' to stop: ").lower()
        if my_choice == 'quit':
            print("Game ended.")
            break
        if my_choice not in choices:
            print("Invalid choice, please try again.")
            continue
        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        if my_choice == computer_choice:
            print("It's a tie!")
        elif (my_choice == 'rock' and computer_choice == 'scissor') or \
             (my_choice == 'scissor' and computer_choice == 'paper') or \
             (my_choice == 'paper' and computer_choice == 'rock'):
            print("You win!")
        else:
            print("Computer wins!")

rock_paper_scissors()
```
